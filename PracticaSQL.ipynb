```python
#Importar librerias
import pandas as pd  #  (Sirve para manipula r nuestros elementos)
import sqlite3    # ( Interfaz para bases de datos SQLite conectarme a un api )

ruta_bd = "/content/drive/MyDrive/Análisis y Visualización de datos/database.sqlite"

```

```python
#Coneccion con bases de datos
conn = sqlite3.connect(ruta_bd)
print ("Conexión exitosa")
#Definir cual tabla queremos llamar ya que pueden haber muchas tablas
tables = pd.read_sql("""SELECT * FROM sqlite_master WHERE type= 'table';""",conn)

tables
```

```python
query = """SELECT * FROM Country; """

countries = pd.read_sql(query, conn)

countries
```

```python
# Consulta todas las ligas del futbol europeo

query = """ SELECT * FROM league;"""

# no modificar lo siguiente
leagues = pd.read_sql(query, conn)
leagues
```

```python
# Consulta de los jugadores en el futbol europeo
query = """ SELECT * FROM Player;"""

Player = pd.read_sql(query, conn)
Player
```

```python
# Se utiliza solo para devolver los valores distintos (no mostrar valores por duplicados)

query = """ SELECT DISTINCT player_name FROM Player;"""

Distinct = pd.read_sql(query, conn)

Distinct
```

```python
# Listar las temporadas jugadas

query = """ SELECT season FROM Match"""

season = pd.read_sql(query, conn)

season
```

```python
# Seleccionar solo los valores distintos de las temporadas

query = """ SELECT DISTINCT season FROM Match"""

seasons = pd.read_sql(query, conn)
seasons
```

```python
# Listar solo las tasas diferentes de valoracion general de los jugadores

query = """ SELECT DISTINCT overall_rating FROM Player_Attributes  """

overall_rating = pd.read_sql(query, conn)
overall_rating
```

```python
# Limitar la cantidad de registros que queremos conocer

query = """SELECT * FROM Player LIMIT 10"""

Limit_Player = pd.read_sql(query, conn)

Limit_Player
```

```python
# Listar 15 jugadores de la base de datos del futbol europeo

query = """ SELECT id, player_name, birthday, height, weight FROM Player LIMIT 15"""

Player_15 = pd.read_sql(query, conn)
Player_15
```

```python
# Listar 25 equipos del futbol europeo

query = """ SELECT id, team_long_name FROM team LIMIT 25"""

Team_25 = pd.read_sql(query, conn)
Team_25
```

```python
# ¿Cuantos Paises diferentes están almacenados en la tabla Country?

query = """ SELECT COUNT(DISTINCT Name) FROM Country"""

count_country = pd.read_sql(query, conn)

count_country
```

```python
# ¿Cuantas etapas diferentes se jugaron?

query = """SELECT COUNT(DISTINCT stage) FROM Match"""

stage_distinct = pd.read_sql(query, conn)

stage_distinct
```

```python
# ¿Cual fue el total de goles hechos por los equipos visitantes?

query = """ SELECT SUM(away_team_goal) FROM Match"""

Goals_away_team = pd.read_sql(query, conn)

Goals_away_team
```

```python
# Promedio de goles hechos por los equipos locales y el promedio de goles hechos por los visitantes de la tabla Match

query= """SELECT AVG(home_team_goal), AVG(away_team_goal) FROM Match"""

Average = pd.read_sql(query, conn)

Average
```

```python
# Máxima cantidad de goles que un equipo visitante anotó en los juegos de la tabla Match

query= """ SELECT MAX(away_team_goal) from MATCH"""

Value_max_goal = pd.read_sql(query, conn)

Value_max_goal
```

```python
# query= """ SELECT MAX(away_team_goal) from MATCH"""

query= """ SELECT * from Country"""

MATCH = pd.read_sql(query, conn)

MATCH
```

```python
# Traer el promedio de goles por partido de la tabla Match

query = """ SELECT AVG(home_team_goal + away_team_goal) FROM Match"""

Promedio_goles = pd.read_sql(query, conn)

Promedio_goles
```

```python
# Jugadores con peso de 165kg

query = """SELECT id,player_name, birthday, height, weight FROM Player WHERE weight=165 LIMIT 15"""

Player_165kg = pd.read_sql(query, conn)

Player_165kg
```

```python
# Listar las temporadas en que un equipo visitante anotó 7 o más goles

query = """ SELECT season, home_team_goal, away_team_goal FROM Match WHERE away_team_goal >= 7"""

Away_goal = pd.read_sql(query, conn)

Away_goal
```

```python
# Traer el Id de los equipos que desarrollan un dribbling normal y su velocidad de juego es balanceada

query = """ SELECT id, buildUpPlayDribblingClass, buildUpPlaySpeedClass FROM Team_Attributes WHERE buildUpPlayDribblingClass = 'Normal' AND buildUpPlaySpeedClass ='Balanced'"""

Team_normal_balanced = pd.read_sql(query, conn)

Team_normal_balanced
```

```python
# Id de los equipos que desarrollan un juego de pase largo o su velocidad de juego es rápida

query= """ SELECT id, buildUpPlayPassingClass, buildUpPlaySpeedClass FROM Team_Attributes WHERE buildUpPlayPassingClass = 'Long' OR buildUpPlaySpeedClass = 'Fast'"""

team_att_or = pd.read_sql(query, conn)

team_att_or
```

```python
# Traer la tabla de países sin Bélgica

query = """SELECT id, name FROM Country WHERE NOT 'Belgium'"""

countries_not = pd.read_sql(query, conn)

countries_not
```

```python
#  Tabla de países únicamente con Italia

query= """ SELECT id, name FROM Country WHERE name = 'Italy' """

countries_equal = pd.read_sql (query, conn)
countries_equal
```

```python
# Traer la tabla de países sin Polonia

query= """ SELECT id, name FROM Country WHERE name != 'Poland' """

countries_dif = pd.read_sql (query, conn)
countries_dif
```

```python
# Listar los jugadores que pesen entre 174 y 176 libras

query= """ SELECT id, player_name, birthday, height, weight FROM Player WHERE weight BETWEEN 174 AND 176 LIMIT 30 """

players_weight = pd.read_sql(query, conn)
players_weight
```

```python
# Listar a los Messi de la base de datos de jugadores

query = """ SELECT id,player_name, birthday, height, weight FROM Player WHERE player_name LIKE '%Messi%' """

players_messi = pd.read_sql(query, conn)
players_messi
```

```python
# Listar a los equipos y ordenarlos por nombre

query = """ SELECT team_long_name FROM Team ORDER BY team_long_name ASC LIMIT 30"""

teams_orderby = pd.read_sql(query, conn)
teams_orderby
```

```python
# Listar la tabla de países en orden descendente

query = """ SELECT id, name FROM Country ORDER BY name DESC"""

teams_orderdesc = pd.read_sql(query, conn)
teams_orderdesc
```

```python
# Listar el id de los 15 jugadores con menos peso

query = """ SELECT id, player_name, weight FROM Player ORDER BY weight ASC LIMIT 15"""

player_weightasc = pd.read_sql(query, conn)
player_weightasc
```

```python
# Listar el id de los 10 jugadores con mayor estatura

query = """ SELECT id, player_name, height FROM Player ORDER BY height DESC LIMIT 10"""

player_heightdesc = pd.read_sql(query, conn)
player_heightdesc
```

```python
# Listar el id de los 10 jugadores con mayor estatura y ordenarlos alfabéticamente pero de forma descendente

query = """ SELECT id, player_name, height FROM Player ORDER BY height DESC, player_name DESC LIMIT 10"""

player_height_name = pd.read_sql(query, conn)
player_height_name
```

```python
# Cuantos equipos tienen cada tipo de velocidad de juego

query = """ SELECT buildUpPlaySpeedClass, COUNT(id) FROM Team_Attributes GROUP BY buildUpPlaySpeedClass = 'Fast', buildUpPlaySpeedClass = 'Balanced', buildUpPlaySpeedClass = 'Slow' """

teams_speed = pd.read_sql(query, conn)
teams_speed
```
