```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lectura
from google.colab import drive
drive.mount('/content/drive')

ruta = "/content/drive/MyDrive/Colab Notebooks/dataset_banco.csv"


data = pd.read_csv(ruta)
print()
```

```python
# Eliminar valores faltantes

data.isna().sum()
```

```python
data.dropna(inplace=True)
data.info()
```

```python
# Conteo de los niveles en en las diferentes columnas categoricas

cols_cat = ['job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign','pdays','previous', 'poutcome', 'y']

for col in cols_cat:
   print (f'Columna.{col}:{data[col].nunique()} subniveles')
```

```python
# Coenteo de valores con -1
count = data['pdays'].value_counts()[-1]
print(count)
```

```python
# Calcular la media del atributo "pdays': dias de contacto al usuario de la campaña actual"

media= data['pdays'].mean()

print("Promedio de días de contacto", media)
```

```python
# Calcular la desviación estandar del atributo "pdays" : dias de contacto al usuario de la campaña actual"

desviacion_estandar = data['pdays'].std()
print("Desviacion estandar de la variable pdays:", desviacion_estandar)
```

```python
#Graficar un histograma del atributo "pdays"
plt.hist(data['pdays'], bins=20, color= 'skyblue', edgecolor='black', alpha=0.5)

#Agregar una linea vertical para mostrar la media
plt.axvline(media, color= 'red', linestyle= 'dashed', linewidth=1)


#Etiquetas y título
plt.xlabel('Valor de la pdays')
plt.ylabel('Frecuencia')
plt.title('Histograma de la pdays con Desviación estandar')
```

```python
# Eliminar filas repetidas

print(f'Tamaño del set antes del aliminar las filas repetidas: {data.shape}')
data.drop_duplicates(inplace=True)
print(f'Tamaño del set despues de aliminar las filas repetidas: {data.shape}')
```

```python
# Generar gráficas indivuiduales pues las variables numericas estan en rngos diferentes

cols_num = ['age', 'balance', 'day', 'duration', 'campaign',
            'pdays', 'previous']

fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(cols_num):
    sns.boxplot(x=col, data=data, ax=ax[i])
    ax[i].set_title(col)
```

```python
# Eliminar filas con "age" >100

print(f' Tamaño del set antes de eliminar los registros de edad: {data.shape}')
data = data.drop(data[data['age'] > 100]. index)
print(f' Tamaño del set despues de eliminar los registros de edad: { data.shape}')
```

```python
# Eliminar filas con " duration" <0

print(f'Tamaño del set antes de eliminar registros de duration:{data.shape}')
data = data.drop(data[data['duration'] < 0].index)
print(f'Tamaño del set despues de eliminar los registros: {data.shape}')
```

```python
# Eliminar filas con "previous" > 100

print(f'Tamaño del set antes de eliminar los registros de previous {data.shape}')
data = data.drop(data[data['duration'] > 100].index)
print(f'Tamaño del set despues de eliminar registros de previous: {data.shape}')
```

Errores tipográficos en variables categóricas


```python
# Gráficar los subniveles de cada variable categórica

cols_cat = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y' ]

fig, ax = plt.subplots (nrows= 10, ncols=1, figsize =(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate (cols_cat):
  sns.countplot(x=col, hue=col, data=data, ax=ax[i], palette= 'colorblind', legend= False)
  ax[i].set_title(col)
  ax[i].tick_params(axis= 'x', labelrotation= 30)
```

# Unificar subniveles  

```python
for column in data.columns:
  # Representar en minusculas solo si la columna es categórica
  if column in cols_cat:
    data[column] = data[column].str.lower()

# Y generemos las gráficas nuevamente
fig, ax = plt. subplots(nrows=10, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate (cols_cat):
  sns.countplot(x=col, hue=col, data=data, ax=ax[i], palette= 'colorblind', legend= False)
  ax[i].set_title(col)
  ax[i].tick_params(axis= 'x', labelrotation= 30)
```

```python
# Unificar admin. y administrative
print(data['job'].unique())
data['job'] = data['job'].str.replace('admin.','administrative', regex= False)
print (data['job'].unique())
```

```python
# marital: Unificar div. y divorced
print(data['marital'].unique())
data['marital'] = data['marital'].str.replace('div.','divorced', regex= False)
print (data['marital'].unique())
```

```python
#  education: unificar sec. y secondary, unk y unknown
print(data['education'].unique())
data['education'] = data['education'].str.replace('sec.','secondary', regex= False)
data.loc[data['education']=='unk''education'] = 'unknown'
print (data['education'].unique())
```

```python
# contact: unificar tephone y phone
print(data['contact'].unique())
data.loc[data['contact']=='phone''contact'] = 'telephone'
data.loc[data['contact']=='mobile''contact'] = 'cellular'

print (data['contact'].unique())
```
